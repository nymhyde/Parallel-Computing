```python
from decimal import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
getcontext().prec = 15
```

```python
index = np.arange(1,13)
print(index)
```

```python
# Defining the error (MPI)
error = np.array([0.0008333314113056,
                  0.0000083333333309,
                  0.0000000833333331,
                  0.0000000008333312,
                  0.0000000000083413,
                  0.0000000000001048,
                  0.0000000000000226,
                  0.0000000000001208,
                  0.0000000000000884,
                  0.0000000013865686,
                  0.0000000017674591,
                  0.0000000002392655])

print(error)
```

```python
# Wall Time (MPI)
time_mpi = np.array([0.000046,
                     0.000042,
                     0.000056,
                     0.000160,
                     0.000143,
                     0.001248,
                     0.010694,
                     0.111374,
                     0.735355,
                     6.882624,
                     67.974126,
                     702.340895])

print(time_mpi)
```

```python
# Wall Time (single process)

wall_single = np.array([0.000026,
                        0.000006,
                        0.000033,
                        0.000035,
                        0.000839,
                        0.003061,
                        0.056281,
                        0.306743,
                        3.078677,
                        32.805832,
                        307.079831,
                        3121.972127])

print(wall_single)
```

```python
# Get Accuracy from Error
acc = []
for i in range(len(error)):
   acc.append(Decimal((1-error[i]))*100)

accuracy = np.array(acc)

print(accuracy)
```

```python
# Plots
#
csfont = {'fontname': 'monospace',
          'weight'  : 'bold',
          'size'    : 15}
hfont = {'fontname': 'monospace',
         'size'    : 15}


# 1. Accuracy vs 10^T interval size

plt.figure(figsize=(8,6), tight_layout=True)
plt.plot(accuracy, linewidth=4, label='Accuracy (in %)')

plt.grid(axis='both',linestyle='-', linewidth=1.5)
plt.xlabel('10^T interval size', **hfont)
plt.ylabel('Accuracy (in %)', **hfont)
plt.title('Effect of Interval Size (10^T) on Accuracy of 'r'$\pi$', **csfont)
plt.legend(loc='best')

plt.savefig("accuracy.pdf", format="pdf", bbox_inches="tight")
```

```python
# 2. Error vs 10^T interval size

plt.figure(figsize=(8,6), tight_layout=True)
plt.plot(error, linewidth=4, label='Absolute Error')

plt.grid(axis='both',linestyle='-', linewidth=1.5)
plt.xlabel('10^T interval size', **hfont)
plt.ylabel('Error', **hfont)
plt.title('Effect of Interval Size (10^T) on Error in 'r'$\pi$', **csfont)
plt.legend(loc='best')

plt.savefig("error.pdf", format="pdf", bbox_inches="tight")
```

```python
# 3. Wall Time Comparison vs 10^T interval size

plt.figure(figsize=(8,6), tight_layout=True)
plt.plot(index, time_mpi, linewidth=4, label='MPI Wall Time (in seconds)')
plt.plot(index, wall_single, linewidth=4, label='Single Process Wall Time (in seconds)')

plt.grid(axis='both',linestyle='-', linewidth=1.5)
plt.xlabel('10^T interval size', **hfont)
plt.ylabel('Wall Time (in seconds)', **hfont)
plt.title('Effect of Interval Size (10^T) on Wall Time of program', **csfont)
plt.legend(loc='best')

plt.savefig("walltime.pdf", format="pdf", bbox_inches="tight")
```

```python
# 4. Wall Time Comparison vs 10^T interval size

plt.figure(figsize=(8,6), tight_layout=True)
plt.plot(index[:5], time_mpi[:5], linewidth=4, label='MPI Wall Time (in seconds)')
plt.plot(index[:5], wall_single[:5], linewidth=4, label='Single Process Wall Time (in seconds)')

plt.grid(axis='both',linestyle='-', linewidth=1.5)
plt.xlabel('10^T interval size', **hfont)
plt.ylabel('Wall Time (in seconds)', **hfont)
plt.title('Effect of Interval Size (10^T) on Wall Time of program', **csfont)
plt.legend(loc='best')

plt.savefig("walltime (zoomed).pdf", format="pdf", bbox_inches="tight")
```
